## Loan Risk Analysis — Applicant Credit Assessment

This notebook covers:
- Loading and inspecting the loan risk dataset (`loan_risk_data.csv`)
- Understanding schema, data types, and distributions
- Basic data quality checks (nulls, duplicates, outliers)
- Saving a cleaned copy for downstream analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')

os.makedirs('data', exist_ok=True)
os.makedirs('outputs', exist_ok=True)

print('✅ Libraries loaded')
```

## 1.1 Load Dataset

```python
df = pd.read_csv('loan_risk_data.csv', parse_dates=['application_date'])

print(f'✅ Dataset loaded: {df.shape[0]:,} rows × {df.shape[1]} columns')
print(f'Date range: {df["application_date"].min().date()} → {df["application_date"].max().date()}')
df.head()
```

## 1.2 Schema & Data Types

```python
print('=== Column Types ===')
print(df.dtypes)
print()
print('=== Column Categories ===')
categorical_cols = df.select_dtypes(include='object').columns.tolist()
numerical_cols   = df.select_dtypes(include=[np.number]).columns.tolist()
print(f'Categorical ({len(categorical_cols)}): {categorical_cols}')
print(f'Numerical   ({len(numerical_cols)}): {numerical_cols}')
```

## 1.3 Target Variable Overview

```python
print('=== Risk Level Distribution ===')
print(df['risk_level'].value_counts())
print(df['risk_level'].value_counts(normalize=True).mul(100).round(2).astype(str) + '%')

print('\n=== Approval Status Distribution ===')
print(df['approval_status'].value_counts())

print('\n=== Risk × Approval Crosstab ===')
print(pd.crosstab(df['risk_level'], df['approval_status'], margins=True))
```

## 1.4 Key Feature Distributions

```python
print('=== Loan Purpose ===')
print(df['loan_purpose'].value_counts())

print('\n=== Employment Type ===')
print(df['employment_type'].value_counts())

print('\n=== Top States ===')
print(df['state'].value_counts().head(10))

print('\n=== Education Level ===')
print(df['education_level'].value_counts())
```

## 1.5 Numerical Summary Statistics

```python
key_nums = ['age', 'annual_income_inr', 'loan_amount_inr', 'credit_score',
            'debt_to_income_ratio', 'savings_balance_inr', 'interest_rate_pct',
            'years_employed', 'loan_tenure_months']

print('=== Summary Statistics ===')
df[key_nums].describe().round(2)
```

## 1.6 Data Quality Check

```python
print('=== Missing Values ===')
nulls = df.isnull().sum()
print(nulls[nulls > 0] if nulls.any() else 'No missing values ✅')

print('\n=== Duplicate Rows ===')
dups = df.duplicated().sum()
print(f'{dups} duplicate rows' if dups > 0 else 'No duplicates ✅')

print('\n=== Unique Applicant IDs ===')
print(f'{df["applicant_id"].nunique()} unique IDs out of {len(df)} rows')

print('\n=== Credit Score Range ===')
print(f'Min: {df["credit_score"].min()} | Max: {df["credit_score"].max()} | Mean: {df["credit_score"].mean():.0f}')

print('\n=== Zero / Negative Collateral Values ===')
zero_collateral = (df['collateral_value_inr'] == 0).sum()
print(f'{zero_collateral} applicants with zero collateral (no collateral pledged)')
```

## 1.7 Feature Engineering — Derived Columns

```python
# Income-to-loan ratio
df['income_to_loan_ratio'] = (df['annual_income_inr'] / df['loan_amount_inr']).round(4)

# Savings-to-expense ratio (months of savings)
df['savings_months_cover'] = (df['savings_balance_inr'] / df['monthly_expenses_inr'].replace(0, np.nan)).round(2)

# Age group
df['age_group'] = pd.cut(df['age'], bins=[20, 30, 40, 50, 65],
                          labels=['21–30', '31–40', '41–50', '51–64'])

# Application month
df['application_month'] = df['application_date'].dt.to_period('M').astype(str)

print('✅ Derived features added:')
print('  - income_to_loan_ratio')
print('  - savings_months_cover')
print('  - age_group')
print('  - application_month')
df[['applicant_id', 'income_to_loan_ratio', 'savings_months_cover', 'age_group', 'application_month']].head()
```

## 1.8 Save Enriched Dataset

```python
df.to_csv('data/loan_risk_enriched.csv', index=False)

print(f'✅ Enriched dataset saved → data/loan_risk_enriched.csv')
print(f'   Shape: {df.shape[0]:,} rows × {df.shape[1]} columns')
print()
print('=== Final Column List ===')
for i, col in enumerate(df.columns, 1):
    print(f'  {i:02d}. {col}')
```
